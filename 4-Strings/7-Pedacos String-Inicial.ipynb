### Forma Básica

```python
precos = "Jan: 25, Fev: 27, Mar: 29"

preco_jan= precos[5:7]
print(preco_jan)
```

### Posição Inicial e Final

```python
preco_mar= precos[-2:]
print(preco_mar)

preco_fev=precos[14:16]
print(preco_fev)

# Outra forma de contegm de trás pra frente

preco_fev2=precos[-11:-9]
print(preco_fev2)
```

### Posição Inicial e Final com Step

```python
codigo = "1.2.3.4,5,1,2.3.4,7.9"

pedaco_cod=codigo[::2]
print(pedaco_cod)
```

```python
# outras formas

codigo = "1.2.3.4,5,1,2.3.4,7.9"

num_codigos = codigo[-1::-1]
print(num_codigos)
```
